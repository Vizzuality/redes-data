# Images
## Purpose
The purpose of this notebook is to create images from Earth Engine datasets.

## Setup
### Library import

```python
import os
from ipywidgets import interact
import matplotlib.pyplot as plt

from vizzDL import images
```

## **Create image**
**Create `images` object**

```python
image = images.Composite()
```

**Select region**

```python
geometry = {'type': 'FeatureCollection',
 'features': [{'type': 'Feature',
   'properties': {},
   'geometry': {'type': 'Polygon',
    'coordinates': [[[17.612549, -14.719569],
      [17.612549, -14.599695],
      [17.757939, -14.599695],
      [17.757939, -14.719569],
      [17.612549, -14.719569]]]}}]}
```

```python
#image.select_region(slug="Landsat-8-Surface-Reflectance", init_date='2019-01-01', end_date='2019-12-31', lat=46.056, lon=-122.1149, zoom=10)
image.select_region(slug="SRTM-Digital-Elevation", init_date='2019-01-01', end_date='2019-12-31', lat=46.056, lon=-122.1149, zoom=10)
```

### Image
**Create image as a Numpy array**

```python
array = image.image_as_array()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.imshow(array[0,:,:,:])
ax.set_title('LR')

plt.show()
```
