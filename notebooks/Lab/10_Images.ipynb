# Images
## Purpose
The purpose of this notebook is to create images from Earth Engine datasets.

## Setup
### Library import

```python
import matplotlib.pyplot as plt

from vizzDL import images
```

## **Create image**
**Create `images` object**

```python
image = images.Composite.DEM()
#image = images.Composite.Landsat(init_date='2019-01-01', end_date='2019-12-31')
```

**Select region**

```python
geometry = {'type': 'FeatureCollection',
 'features': [{'type': 'Feature',
   'properties': {},
   'geometry': {'type': 'Polygon',
    'coordinates': [[[-122.206748, 46.034209],
      [-122.206748, 46.206846],
      [-122.010182, 46.206846],
      [-122.010182, 46.034209],
      [-122.206748, 46.034209]]]}}]}
```

```python
image.select_region(geometry=geometry, lat=46.156, lon=-122.1149, zoom=10)
```

### Image

**Save image as `PNG`**

```python
image.save_as_PNG(folder_path = '../../datasets/processed/GeoTIFFs/', region_name = 'Saint_Helens')
```

**Create image as a Numpy array**

```python
array = image.image_as_array()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.imshow(array)

plt.show()
```

**Save image as `GeoTIFF`**

```python
image.save_as_GeoTIFF(folder_path = '../../datasets/processed/GeoTIFFs/', region_name = 'Saint_Helens')
```
